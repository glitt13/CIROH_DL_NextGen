```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Directory containing the .gpkg files
directory = "/home/jmframe/ngen/extern/lstm/hydrofabric/v20.1/gpkg/"

# Initialize a dictionary to store GeoDataFrames with the filename as the key
gpkgs = {}

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".gpkg"):
        filepath = os.path.join(directory, filename)
        try:
            # Read the .gpkg file into a GeoDataFrame
            gdf = gpd.read_file(filepath)
            gpkgs[filename] = gdf  # Store the GeoDataFrame
            print(f"Successfully loaded: {filename}")
        except Exception as e:
            print(f"Failed to load {filename}: {e}")
```

```python
# Print summary of loaded GeoDataFrames, including columns
for name, gdf in gpkgs.items():
    print(f"{name}: {gdf.shape[0]} rows, {gdf.shape[1]} columns")
    print(f"Columns in {name}: {list(gdf.columns)}\n")

# Example: Inspect the first few rows of one GeoDataFrame (e.g., nextgen_01.gpkg)
example_gdf = gpkgs["nextgen_01.gpkg"]
print("Example rows from nextgen_01.gpkg:")
print(example_gdf.head())
```

```python
# Directory containing the .gpkg files
directory = "/home/jmframe/ngen/extern/lstm/hydrofabric/v20.1/gpkg/"

# Initialize an empty list to store GeoDataFrames
gdfs = []

# Loop through all .gpkg files and load them
for filename in os.listdir(directory):
    if filename.endswith(".gpkg"):
        filepath = os.path.join(directory, filename)
        try:
            # Load the GeoDataFrame
            gdf = gpd.read_file(filepath)
            gdfs.append(gdf)  # Store for plotting
            print(f"Successfully loaded: {filename}")
        except Exception as e:
            print(f"Failed to load {filename}: {e}")

# Plot all geometries on the same graph
plt.figure(figsize=(10, 10))

for gdf in gdfs:
    gdf.plot(ax=plt.gca(), alpha=0.6, linewidth=0.5, edgecolor=None)

# Add title and grid
plt.title("Combined Plot of All GeoPackage Files")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define paths
vpu = "vpu07"  # Example VPU to process
output_dir = f"/home/jmframe/ngen/extern/lstm/ngen_output/{vpu}"
gpkg_path = f"/home/jmframe/ngen/extern/lstm/hydrofabric/v20.1/gpkg/nextgen_07.gpkg"

# Load the GeoPackage
gdf = gpd.read_file(gpkg_path)

# Initialize a dictionary to store the last nexus output values
nexus_values = {}

# Loop through each file in the VPU directory that matches the pattern 'nex-*_output.csv'
for filename in os.listdir(output_dir):
    if filename.startswith("nex-") and filename.endswith("_output.csv"):
        # Extract nexus ID from the filename
        nexus_id = filename.split('-')[1].split('_')[0]

        # Read the last line of the CSV file to get the final value
        file_path = os.path.join(output_dir, filename)
        df = pd.read_csv(file_path, header=None)

        # Extract the last value from the 3rd column
        last_value = df.iloc[-1, 2]

        # Store the value with the corresponding nexus ID
        nexus_values[nexus_id] = last_value

# Convert the dictionary to a DataFrame for merging with the GeoPackage
df_nexus_values = pd.DataFrame(list(nexus_values.items()), columns=["id", "last_value"])
df_nexus_values["id"] = "nex-" + df_nexus_values["id"]  # Add 'nex-' prefix to match GeoPackage IDs

# Merge GeoPackage data with the nexus output values
gdf_merged = gdf.merge(df_nexus_values, on="id", how="left")

# Plot the GeoPackage points, colored by the last nexus output values
fig, ax = plt.subplots(figsize=(10, 8))
gdf_merged.plot(
    column="last_value",  # Use the last value for color mapping
    cmap="coolwarm",  # Colormap for visualization
    legend=True,
    markersize=20,
    ax=ax
)

# Add labels and title
plt.title(f"Nexus Points for {vpu}, Colored by Last Output Value")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.show()
```
